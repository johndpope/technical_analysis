```python
import pandas as pd
import numpy as np

pairs = ['EURJPY', 'EURUSD', 'USDJPY']
times = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007',
         '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015',
         '2016', '2017', '201801', '201801']

for pair in pairs[:1]:
    print('Processing %s...' % pair)
    file_stem = 'fx_data/' + pair + '/DAT_ASCII_' + pair + '_M1_'
    if pair == 'EURJPY':
        year_range = times[2:]
    else:
        year_range = times

    df = pd.read_csv(file_stem + year_range[0] + '.csv', sep=';', header=None)
    for year in year_range[1:2]:
        new = pd.read_csv(file_stem + year + '.csv', sep=';', header=None)
        df = pd.concat([df, new], axis=0)
    
    df.columns = ['time', 'open', 'high', 'low', 'close', 'volume']
    df['time'] = pd.to_datetime(df['time'], infer_datetime_format=True)
    df = df.set_index('time')
    del df['volume']
```
