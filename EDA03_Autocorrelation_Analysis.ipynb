# 1. USD/JPY Pair 2000-2018_Feb

```python
if True:
    # Import USD/JPY hourly data
    data = pd.read_csv('fx_data/USDJPY/USDJPY_2000-201802_hour.csv', header=0)
    data['time'] = pd.to_datetime(data['time'], infer_datetime_format=True)
```

```python
ta = None
r_periods = [1, 2, 6, 24, 120, 240, 480, 2400]
ta = base.TechnicalAnalysisCase(data)
ta.add_returns(r_periods)
```

```python
ta.data.describe()
```

```python
plus_labels = [col for col in ta.data.columns if 'return' in col and '+' in col]
minus_labels = [col for col in ta.data.columns if 'return' in col and '-' in col]
return_labels = minus_labels + plus_labels
print(plus_labels)
print(minus_labels)
```

```python
print('*'*10, "Auto-correlation Test", '*'*10)
fig1, ax1 = plt.subplots(1, 1, figsize=(12, 2))
autocorrelation_plot(ta.data.close, color='salmon', ax=ax1, label='close')
ax1.set_title('Auto-correlation of the closing price')

fig2 = plt.figure(figsize=(12, 8))
for i, label in enumerate(plus_labels):
    ax = fig2.add_subplot(4, 2, i+1)
    ax.axvline(r_periods[i], ls='--', color='0.7')
    autocorrelation_plot(ta.data[label].dropna(), color='salmon', ax=ax, label=label)
    ax.set_xlabel('Lag (Hour)')
    ax.set_ylabel('Auto-orrelation')
    #ax.legend()
    ax.grid(False)
    #ax.set_xlim([-100, 5000])
    
plt.tight_layout()
```

```python
# Lag plot
fig2 = plt.figure(figsize=(12, 6))
for i, label in enumerate(plus_labels):
    ax = fig2.add_subplot(2, 4, i+1)
    lag_plot(ta.data[label].dropna(), lag=r_periods[i], c='skyblue', s=5, alpha=0.5, ax=ax, label=label)
    ax.legend()
    ax.grid(False)
    
plt.tight_layout()
```
